En esta parte se encuentra codigo de los m√≠nimos cuadrados y con incertidumbre

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])

# Generar valores de y (con ruido gaussiano)
y = 2 * x + 1 + np.random.normal(0, 1, len(x))

# Generar valores de incertidumbre (pueden variar para cada punto)
i = np.array([0.5, 0.6, 0.7, 0.8, 0.9])

# Graficar los puntos con barras de error
plt.errorbar(x, y, yerr=i, fmt='o', label='Datos con incertidumbre')

def minimos_cuadrados(x,y):
    n = len(x)
    sum_x = np.sum(x)
    sum_y = np.sum(y)
    sum_xy = np.sum(x*y)
    sum_x_2 = np.sum(x**2)
    pendiente = (n*sum_xy-(sum_x*sum_y))/(n*sum_x_2-(abs(sum_x)**2))
    intercepto = (sum_y*sum_x_2-sum_x*np.sum(x*y))/(n*sum_x_2-(abs(sum_x)**2))
    return pendiente,intercepto

m,b = minimos_cuadrados(x,y)
print(m,b)
# A√±adir etiquetas y t√≠tulo
plt.plot(x,m*x+b)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresi√≥n Lineal con Incertidumbre')
plt.legend()

# Mostrar la gr√°fica
plt.show()
```
